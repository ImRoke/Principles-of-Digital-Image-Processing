```python
#import the required modules
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading and resizing images
img1 = cv.imread('goku.png')
img1 = cv.resize(img1, (300, 300))
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
```

```python
#Cropping image by slicing technique
img_crop = img1[0:150, 100:200]
plt.imshow(cv.cvtColor(img_crop, cv.COLOR_BGR2RGB))
```

```python
#Horizantally flipping the image by using default method
img_flip = cv.flip(img1, 0)
plt.imshow(cv.cvtColor(img_flip, cv.COLOR_BGR2RGB))
```

```python
#getting to know the available rotate options
rot = [i for i in dir(cv) if i.startswith('ROTATE_')]
print("No of available rotations are: {}".format(len(rot)))
print(rot)
```

```python
#Rotating the image 180 degrees clockwise by using default method
img_rot_180 = cv.rotate(img1, cv.ROTATE_180)
plt.imshow(cv.cvtColor(img_rot_180, cv.COLOR_BGR2RGB))
```

```python
#Rotating the image 90 degrees clockwise by using default method
img_rot_90cw = cv.rotate(img1, cv.ROTATE_90_CLOCKWISE)
plt.imshow(cv.cvtColor(img_rot_90cw, cv.COLOR_BGR2RGB))
```

```python
#Rotating the image 90 degrees Anti-clockwise by using default method
img_rot_90acw = cv.rotate(img1, cv.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(cv.cvtColor(img_rot_90acw, cv.COLOR_BGR2RGB))
```

```python
#Scratch implementation of rotating 90 acw direction

img3 = img1.copy()
img3[:,:,:] = [0, 0, 0]

for i in range(300):
    for j in range(300):
        img3[i,j]=img1[j,i]
        
plt.imshow(cv.cvtColor(img3, cv.COLOR_BGR2RGB))        
```

```python
#Scratch implementation of rotating 180 cw direction

#creating an empty image with all the values equals to zero
img4 = img1.copy()
img4[:,:,:] = [0, 0, 0]

#iterating through the height and width of the image
for i in range(300):
    for j in range(300):
        
        #merging the pixel values with empty image
        img4[i,j]=img1[299-i,299-j]
        
plt.imshow(cv.cvtColor(img4, cv.COLOR_BGR2RGB)) 
```

```python
#Scratch implementation of rotating 90 cw direction

#creating an empty image with all the values equals to zero
img5 = img1.copy()
img5[:,:,:] = [0, 0, 0]

#iterating through the height and width of the image
for i in range(300):
    for j in range(300):
        
        #changing the axis and then merging the pixel values with empty image
        img5[j,i]=img1[299-i,299-j]
        
plt.imshow(cv.cvtColor(img5, cv.COLOR_BGR2RGB))
```

```python
#Plotting all the results on a grid

titles = ["Original Image","Flipped Horizontally","Rotated_180", "Rotated_180_Scratch","Rotated_90CW","Rotated_90_CW_Scratch", "Rotated_90CCW", "Rotated_90_CCW_Scratch"]
images = [img1, img_flip, img_rot_180, img4, img_rot_90cw, img5, img_rot_90acw, img3]
plt.figure(figsize=(25,15))
for i in range(8):
    plt.subplot(4,2,i+1)
    plt.imshow(cv.cvtColor(images[i], cv.COLOR_BGR2RGB))
    plt.title(titles[i])
plt.tight_layout()
plt.show()
```

Brightness and Contrast Control:

output = (gain_contrats * image) + bias_brightness

new_image(i,j) = alpha*image(i,j) + beta

alpha 1  beta 0      --> no change  
0 < alpha < 1        --> lower contrast  
alpha > 1            --> higher contrast  
-127 < beta < +127   --> good range for brightness values


```python
#Brightness and Contrast Control

#creating an empty image with all the values equals to zero
out = img1.copy()
out[:,:,:] = [0,0,0]

#iterating through all the dimesions of image
for ii in range(img1.shape[0]):
    for jj in range(img1.shape[1]):
        for kk in range(img1.shape[2]):
            
            #the output follows as (gain_contrats * image) + bias_brightness
            out[ii, jj, kk] = 0.5 * img1[ii,jj,kk] + 127
            
plt.imshow(cv.cvtColor(out, cv.COLOR_BGR2RGB))
```

[x             y]=[cos()sin()             sin()cos()][x             y]

```python
# custom rotation of image at a particular value

#creating an empty image with all the values equals to zero
img7 = np.zeros((img1.shape), dtype = np.uint8)
h, w = img7.shape[0], img7.shape[1]

theta = 245
radians = np.radians(theta)

# iterating through the height and width of the image
for i in range(h):
    for j in range(w):
        
        # rotating the image over the image by calculating centre points
        # calculating shifted image pixel value by using orthogonal matrix
        rot_x =  round((i - w//2 ) * np.cos(radians) - (j - h//2) * np.sin(radians)) + w//2
        rot_y = round((i - w//2 ) * np.sin(radians) + (j - h//2) * np.cos(radians)) + h//2
        
        
        # checking the condition for centre points should be greater or equal to zero and 
        # which are lesser than the dimensions of original image
        if (rot_x >= 0 and rot_y >= 0 and rot_x < img1.shape[0] and  rot_y < img1.shape[1]):
            
            #changing the axis and then merging the pixel values with empty image
            img7[i, j] = img1[rot_x, rot_y]
            
plt.imshow(cv.cvtColor(img7, cv.COLOR_BGR2RGB))
plt.title("Custom Rotation of {}".format(theta))
```
